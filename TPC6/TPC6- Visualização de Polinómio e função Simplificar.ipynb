```python
import matplotlib.pyplot as plt

#cria termo do polinomio em tuplo
def criarTermo():
    coef=int(input("Valor do coeficiente: "))
    exp=int(input("Respetivo expoente: "))
    return (coef,exp)

#permite ver termo na forma "correta"
def verTermo(t):
    c,e=t
    if e==0:
        return(str(c))
    elif c==1:
        return ("x^"+str(e))
    else:
        return (str(c)+"x^"+str(e))

#lista com tuplos dos termos do polinomio
def criarPolinomio():
    p=[]
    t=int(input("Quantos termos tem o seu polinómio?: "))
    for i in range(t):
        p.append(criarTermo())
    return p
   
#ver expressão do polinómio 
def verPolinomio(p):
    politexto=""
    if len(p)!=0:
        politexto=verTermo(p[0])
        for i in range(1,len(p)):
            if p[i][0]>0:
                politexto= politexto + "+" + verTermo(p[i])
            else:
                politexto= politexto + verTermo(p[i])
    return (politexto)

#ver grau do polinomio
def verGrau(p):
    grau=p[0][1]
    for i in range (len(p)):
        if p[i][1]>grau:
            grau=p[i][1]
    return grau

#derivada polinomio
def derivada(p):
    derivada=[]
    for (c,e) in p:
        if e!=0:
          derivada.append((c*e, e-1))
    return derivada
    
def calcPol(p,x):
    valor=0
    for (c,e) in p:
        valor= valor + (c*x)**e
    return valor

#simplificar polinomio
def ordenar_l(p):
    paux=[]
    i=verGrau(p)
    while i>=0:
        for(c,e) in p:
            if e==i:
                paux.append((c,e))
        i=i-1
    return paux

def simplificaPol(p):
    paux=ordenar_l(p)
    simplificada=[]
    while len(paux)>1:
        res=paux[0]
        del paux[0]
        while (res[1]==paux[0][1] and len(paux)>1):
            res=(res[0]+paux[0][0],res[1])
            del paux[0]
        simplificada.append(res)
    
    if paux[0][1]==simplificada[-1][1]:
       res=(paux[0][0]+simplificada[-1][0],paux[0][1])
       del simplificada[-1]
       simplificada.append(res)
    else:
       simplificada.append(res)
        
    return simplificada
        

def VisualizarPolinomio(pol):
    x=[n for n in range(-100,100)]
    y=[]
    for i in x:
        y.append(calcPol(pol,i))
    eixoXx = [x[0]-1,x[-1]+1]
    eixoXy = [0,0]
    plt.plot(eixoXx, eixoXy, color='red', linewidth = 1)

    eixoYx = [0,0]
    eixoYy = [-200,200]
    plt.plot(eixoYx, eixoYy, color='red', linewidth = 1)
              
    plt.plot(x,y)
    plt.title('Polinómio')
    plt.show()
    

escolha=1
while escolha!=0:
    print(""""Aplicação para poliómios
    0-Sair
    1-Criar Polinómio
    2-Listar Polinómio
    3-Calcular Grau
    4-Calcular valor num ponto
    5-Calcular Derivada
    6-Simplificar Polinómio
    7-Visualizar Polinómio""")
    escolha=int(input("Opção: "))
    if escolha==1:
        Pol=criarPolinomio()
    if escolha==2:
        s=verPolinomio(Pol)
        print (s)
    if escolha==3:
        print ("O grau do polinómio é: ", verGrau(Pol))
    if escolha==4:
        x=int(input("Introduza o valor de x que deseja calcular: "))
        print("O valor do polinómio em x=",x,"é de: ", calcPol(Pol,x))
    if escolha==5:
        u=verPolinomio(derivada(Pol))
        print(u)
    if escolha==6:
        u=verPolinomio(simplificaPol(Pol))
        print(u)
    if escolha==7:
        VisualizarPolinomio(Pol)
        
else:
    print("Escolha inválida.")
```
