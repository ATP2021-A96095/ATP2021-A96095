```python
#cria termo do polinomio em tuplo
def criarTermo():
    coef=int(input("Valor do coeficiente: "))
    exp=int(input("Respetivo expoente: "))
    return (coef,exp)

#permite ver termo na forma "correta"
def verTermo(t):
    c,e=t
    if e==0:
        return(str(c))
    elif c==1:
        return ("x^"+str(e))
    else:
        return (str(c)+"x^"+str(e))

#lista com tuplos dos termos do polinomio
def criarPolinomio():
    p=[]
    t=int(input("Quantos termos tem o seu polinómio?: "))
    for i in range(t):
        p.append(criarTermo())
    return p
   
#ver expressão do polinómio 
def verPolinomio(p):
    politexto=""
    if len(p)!=0:
        politexto=verTermo(p[0])
        for i in range(1,len(p)):
            if p[i][0]>0:
                politexto= politexto + "+" + verTermo(p[i])
            else:
                politexto= politexto + verTermo(p[i])
    return (politexto)

#ver grau do polinomio
def verGrau(p):
    grau=p[0][1]
    for i in range (len(p)):
        if p[i][1]>grau:
            grau=p[i][1]
    return grau

#derivada polinomio
def derivada(p):
    derivada=[]
    for (c,e) in p:
        if e!=0:
          derivada.append((c*e, e-1))
    return derivada
    
def calcPol(p,x):
    valor=0
    for (c,e) in p:
        valor= valor + (c*x)**e
    return valor

#simplificar polinomio
def ordenar_l(p):
    pordenada=[]
    i=verGrau(p)
    while i>=0:
        for(c,e) in p:
            if e==i:
                pordenada.append((c,e))
        i=i-1
    return pordenada

def simplificaPol(p):
    pordenada=ordenar_l(p)
    simplificada=[]
    while len(pordenada)>1:
        res=pordenada[0]
        del pordenada[0]
        while(res[1]==pordenada[0][1]):
            res=(res[0]+pordenada[0][0],pordenada[0][1])
            resultado=res
            del pordenada[0]
        simplificada.append(resultado)
    if pordenada[0][1]!=simplificada[-1][1]:
        simplificada.append(pordenada[0])
    return simplificada
        

escolha=1
while escolha!=0:
    print(""""Aplicação para poliómios
    0-Sair
    1-Criar Polinómio
    2-Listar Polinómio
    3-Calcular Grau
    4-Calcular valor num ponto
    5-Calcular Derivada
    6-Simplificar Polinómio""")
    escolha=int(input("Opção: "))
    if escolha==1:
        Pol=criarPolinomio()
    if escolha==2:
        s=verPolinomio(Pol)
        print (s)
    if escolha==3:
        print ("O grau do polinómio é: ", verGrau(Pol))
    if escolha==4:
        x=int(input("Introduza o valor de x que deseja calcular: "))
        print("O valor do polinómio em x=",x,"é de: ", calcPol(Pol,x))
    if escolha==5:
        verPolinomio(derivada(Pol))
    if escolha==6:
        verPolinomio(simplificaPol(Pol))
        
else:
    print("Escolha inválida.")
```
