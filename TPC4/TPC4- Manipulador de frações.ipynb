```python
#frações introduzidas
def recebeFrac():
    n=int(input("Introduza o valor do numerador: "))
    d=int(input("Introduza o valor do denominador: "))
    return (n,d)

def nFrac():
    f=[]
    x=int(input("Quantas frações deseja adicionar?: "))
    for u in range (x):
        f.append(recebeFrac())
    return f

#frações na forma "correta"
def verFrac(f):
    n,d=f
    return (str(n) + "/" + str(d))

#maximo divisor comum
def mdc(a,b):
    while b!=0:
        resto=a%b
        a=b
        b=resto
    return a
   
#reduzir fração
def simplificaFrac(f):
    n,d=f
    m=mdc(n,d)
    return(int(n/m),int(d/m))
 
#soma frações
def somaFrac(f1,f2):
    n1=f1[0]
    d1=f1[1]
    n2,d2=f2
    return simplificaFrac((n1*d2 + n2*d1, d2*d1))

def somaVarias():
    l=[]
    l=nFrac()
    s=l[0]
    i=1
    while i < (len(l)):
        s=somaFrac(s,l[i])
        i=i+1
    print ("A soma das frações é {} ".format(verFrac(s)))
 
#subtrai frações
def subFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificaFrac((n1*d2 - n2*d1, d2*d1))

def subVarias():
    l=[]
    l=nFrac()
    s=l[0]
    i=1
    while i < (len(l)):
      s=subFrac(s,l[i])
      i=i+1
    print ("A subtração das frações é: ", verFrac(s))
    
#multiplica frações
def multiFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificaFrac((n1*n2, d1*d2))

def multiVarias():
    l=[]
    l=nFrac()
    s=l[0]
    i=1
    while i < (len(l)):
      s=multiFrac(s,l[i])
      i=i+1
    print ("A multiplicação das frações é: ",verFrac(s))
     
#divide frações
def divideFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificaFrac((n1*d2, n2*d1))

def divideVarias():
    l=[]
    l=nFrac()
    s=l[0]
    i=1
    while i < (len(l)):
      s=divideFrac(s,l[i])
      i=i+1
    print ("A divisão das frações é: ",verFrac(s))
         
print ("Operações com frações")
escolha=1
while escolha!=0:
    print ("1.Soma")
    print ("2.Subtração")
    print ("3.Divisão")
    print ("4.Multiplicação")
    print ("0.Sair")
    escolha=int(input("Opção: "))

    if escolha==1:
        somaVarias()
    if escolha==2:
        subVarias()
    if escolha==3:
        divideVarias()
    if escolha==4:
        multiVarias()
```
