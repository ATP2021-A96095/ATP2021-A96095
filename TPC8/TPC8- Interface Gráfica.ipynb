```python
import codigofunc as fg
import PySimpleGUI as sg

import matplotlib.pyplot as plt

bd=fg.lerDataset("emd.csv")
sg.theme("BluePurple")

layout= [ [sg.Button("Sair", size=(10,1))],
          [sg.Text("")],
          [sg.Text(""), sg.Text(""), sg.Text(""), sg.Button("Listar", size=(10,1)), sg.Button("Distribuições", size=(10,1)), sg.Button("Gráficos", size=(10,1))]
]

window=sg.Window("Exames médicos", size=(400,100)).layout(layout)
stop=False

while not stop:
    event, values= window.read()
    if event == sg.WIN_CLOSED or event == "Sair":
        stop = True
        
    elif event=="Listar":
        layoutListar= [[sg.Button("Listar Modalidades"), sg.Button("Listar por exames")]]
        janelaListar=sg.Window("Opções para Listar", layoutListar)
        escolha1, valores1= janelaListar.read(close=True)
        if escolha1=="Listar Modalidades":
            guardaListaMod=fg.modalidades(bd)
            layoutResultadoMod= [[sg.Text(guardaListaMod), sg.Button("Cancelar")]]
            janelaListaMod=sg.Window("Lista de Modalidades", layoutResultadoMod)
            event1, values1=janelaListaMod.read(close=True)
        if escolha1=="Listar por exames":
            guardaListaEx=fg.listarDataset(bd)
            layoutResultadoEx=[[sg.Listbox(guardaListaEx, size=(100,30)), sg.Button("Cancelar")]]
            janelaListaResul=sg.Window("Lista de Exames", layoutResultadoEx)
            event2, values2=janelaListaResul.read(close=True)
            
    elif event=="Distribuições":
        layoutDistrib= [ [sg.Button("Modalidades")],
                         [sg.Button("Ano")],
                         [sg.Button("Clubes")],
                         [sg.Button("Resultados de exames")] ]
        janelaDistrib=sg.Window("Opções", layoutDistrib)
        escolha2, valores2= janelaDistrib.read(close=True)
        if escolha2 == "Modalidades":
            guardaListaDMod=fg.distrib(bd,6)
            layoutResultadoDMod=[[sg.Text(guardaListaDMod), sg.Button("Cancelar")]]
            janelaDistMod=sg.Window("Distribuição por Modalidades", layoutResultadoDMod)
            event3, values3=janelaDistMod.read(close=True)
        if escolha2 == "Ano":
            guardaListaDAno=fg.distribPorAno1(bd)
            layoutResultadoDAno=[[sg.Text(guardaListaDAno), sg.Button("Cancelar")]]
            janelaDistAno=sg.Window("Distribuição por Ano", layoutResultadoDAno)
            event4, values4=janelaDistAno.read(close=True)
        if escolha2== "Clubes":
            guardaListaDClube=fg.distrib(bd,7)
            layoutResultadoDClube=[[sg.Text(guardaListaDClube), sg.Button("Cancelar")]]
            janelaDistClube=sg.Window("Distribuição por Clubes", layoutResultadoDClube)
            event5, values5=janelaDistClube.read(close=True)
        if escolha2== "Resultados de exames":
            guardaListaDEx=fg.distrib(bd,10)
            layoutResultadoDEx=[[sg.Text("True=Apto; False=Não Apto")],
                                [sg.Text(guardaListaDEx), sg.Button("Cancelar")]]
            janelaDistEx=sg.Window("Distribuição por Exames", layoutResultadoDEx)
            event6, values6=janelaDistEx.read(close=True)
            
    elif event=="Gráficos":
        layoutGraf= [ [sg.Button("Modalidades")],
                      [sg.Button("Ano")],
                      [sg.Button("Clubes")],
                      [sg.Button("Resultados de exames")]
                     ]
        janelaGrafico=sg.Window("Opções", layoutGraf)
        escolha3, valores3= janelaGrafico.read(close=True)
        if escolha3=="Modalidades":
            guardaGrafMod=fg.plotDistrib(bd,6,"Modalidades")
            w=sg.Window("Gráfico da distribuição por modalidades", plt.show(guardaGrafMod))
        if escolha3=="Ano":
            guardaGrafAno=fg.plotDistribPorAno(bd)
            w1=sg.Window("Gráfico da distribuição por ano", plt.show(guardaGrafAno))
        if escolha3=="Clubes":
            guardaGrafClube=fg.plotDistrib(bd,7,"Clubes")
            w2=sg.Window("Gráfico da distribuição por clube", plt.show(guardaGrafClube))
        if escolha3=="Resultados de exames":
            guardaGrafExame=fg.plotDistrib(bd,10,"Exames")
            w3=sg.Window("Gráfico da distribuição por resultado de exames", plt.show(guardaGrafExame))
            
             
window.close()
```
