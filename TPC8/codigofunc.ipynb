```python
def convBool(v):
    return v== "true"

def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    for linha in f:
        novalinha=linha.replace("\n","")
        campos=novalinha.split(",")
        emd=[]
        emd.append("emd"+str(campos[1]))
        emd.append(campos[2])
        emd.append(campos[4] + " " + campos[3])
        emd=emd + campos[5:]
        emd[-1]=convBool(emd[-1])
        emd[-2]=convBool(emd[-2])
        bd.append(emd)
    return bd


def listarDataset(bd):
    l=[]
    for exame in bd:
        if exame[-1]:
            res="Apto"
        else:
            res="Não"
        a= (exame[0] + "|" + exame[2] + "|" + exame[1] + "|" + res)
        l.append(str(a)+"\n")
    return l
    
def modalidades(d):
    modalidades=[]
    for elem in d:
        if elem[6] not in modalidades:
            modalidades.append(elem[6])
    modalidades.sort()   
    return modalidades


def distribPorAno1(bd):
    distribuicao={}
    for elem in bd:
        data=elem[1][0:4]
        if data in distribuicao.keys():
            distribuicao[data]=distribuicao[data]+1
        else:
            distribuicao[data]=1
    return distribuicao

def distrib(bd,index):
    distribuicao={}
    for elem in bd:
        if elem[index] in distribuicao.keys():
            distribuicao[elem[index]]=distribuicao[elem[index]]+1
        else:
            distribuicao[elem[index]]=1
    return distribuicao

def plotDistrib(bd,ind,oque):
    import matplotlib.pyplot as plt
    basedados=distrib(bd,ind)
    left=[]
    height=[]
    tick_label=[]
    i=1
    for chave in basedados:
        left.append(i)
        i=i+3
        height.append(basedados[chave])
        tick_label.append(chave)
    plt.xticks(rotation=45)
    plt.xlabel(oque)
    plt.ylabel("Número")
    plt.bar(left,height,tick_label=tick_label, width=2, color=['teal'])
    

def plotDistribPorAno(bd):
    import matplotlib.pyplot as plt
    left=[]
    height=[]
    tick_label=[]
    a=distribPorAno1(bd)
    i=50
    for elem in a:
        tick_label.append(elem)
        height.append(a[elem])
        left.append(i)
        i=i+1
    plt.bar(left, height, tick_label = tick_label,
        width = 0.8, color = ['red'])
    
    plt.xlabel('Ano')
    plt.ylabel('Número de pessoas')
    plt.title("Distribuição de pessoas por ano")
    
```
