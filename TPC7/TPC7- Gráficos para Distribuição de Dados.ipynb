```python
#converter dados num dicionário
def lerDataset(fnome):
    bd=[]
    f=open( fnome, encoding='utf-8')
    f.readline()
    for linha in f:
        novaLinha=linha.replace("\n","") #elimina \n no fim das linhas
        campos=novaLinha.split(",")
        aluno={}
        aluno["id"]=campos[0]
        aluno["nome"]=campos[1]
        aluno["curso"]=campos[2]
        listaTPC=[] #vai receber os valores do tpc como inteiros
        for tpc in campos[3:]:
            listaTPC.append(int(tpc))
        aluno["tpc"]=listaTPC
        bd.append(aluno)
    return bd

#distribuição de alunos por curso, retorna dicionário com o número total de alunos em cada curso
def distribPorCurso(d):
    distribuicao={}
    for a in d:
        if a['curso'] in distribuicao.keys():
            distribuicao[a['curso']]=distribuicao[a['curso']]+1
        else:
            distribuicao[a['curso']]=1
    return distribuicao
#recebe como parâmetro uma variável com o valor de lerDataset("alunos.csv") ou seja, o dicionário com os dados iniciais

#distribuição de alunos por média, retorna dicionário com o número total de alunos para cada média
def distribPorMedia(d):
    distribuicao={}
    for a in d:
        media= int((sum(a["tpc"])/len(a["tpc"])))
        if media in distribuicao.keys():
            distribuicao[media]=distribuicao[media]+1
        else:
            distribuicao[media]=1
    return distribuicao

#ordenar as médias por ordem crescente, retorna dicionário com as chaves ordenadas
def ordenarDicMedia(d):
    ordenado={}
    for elem in sorted(dict.items(d)) :
        ordenado[elem[0]]=elem[1]
    
    return ordenado
bd=lerDataset("alunos.csv")
d=distribPorMedia(bd)
ordenarDicMedia(d)

#gráfico de barras para a distribuição por curso
import matplotlib.pyplot as plt
def plotDistribPorCurso(bd):
    left = []
    height = []
    tick_label = []
 
    a=distribPorCurso(bd)
    i=1
    for elem in a:
        tick_label.append(elem)
        height.append(a[elem])
        left.append(i)
        i=i+1
    plt.bar(left, height, tick_label = tick_label,
        width = 0.8, color = ['red'])
    
    plt.xlabel('Cursos')
    plt.ylabel('Número de alunos')
    plt.title("Distribuição de alunos por curso")
    plt.show()
bd=lerDataset("alunos.csv")
plotDistribPorCurso(bd)

#gráfico de barras para a distribuição por média
import matplotlib.pyplot as plt
def plotDistribPorMedia(bd):
    left = []
    height = []
    tick_label = []
 
    a=ordenarDicMedia(bd)
    i=1
    for elem in a:
        tick_label.append(elem)
        height.append(a[elem])
        left.append(i)
        i=i+1
    plt.bar(left, height, tick_label = tick_label,
        width = 0.8, color = ['blue'])
    
    plt.xlabel('Média')
    plt.ylabel('Número de alunos')
    plt.title("Distribuição de alunos por média de tpc")
    plt.show()
bd=lerDataset("alunos.csv")
d=distribPorMedia(bd)
plotDistribPorMedia(d)
```
